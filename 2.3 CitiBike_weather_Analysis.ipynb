### 4. Created a column with the count of trips per day and merged it with the weather dataframe.

```python
import pandas as pd  

# Define file paths
weather_path = r"P:\CarrerFoundy\Data Visualizations with Python\Achivement 2\LaGuardia_Weather_2022.csv"
bike_data_path = r"P:\CarrerFoundy\Data Visualizations with Python\Achivement 2\merged_citibike_data_2022.csv"

# Load datasets
weather_df = pd.read_csv(weather_path, parse_dates=["date"])
bike_df = pd.read_csv(bike_data_path, parse_dates=["started_at"])
```

```python
import pandas as pd

# Check the first few rows of 'started_at' to understand its format
print(bike_df["started_at"].head())

# Convert 'started_at' to datetime, handling custom formats
bike_df["started_at"] = pd.to_datetime(bike_df["started_at"], errors="coerce")

# Extract only the date
bike_df["date"] = bike_df["started_at"].dt.date

# Count trips per day
trips_per_day = bike_df.groupby("date").size().reset_index(name="trip_count")

# Ensure the data covers the full 2022 range
all_dates = pd.date_range(start="2022-01-01", end="2022-12-31").date
all_dates_df = pd.DataFrame(all_dates, columns=["date"])

# Merge to include all dates (even those without trips)
trips_per_day_full = pd.merge(all_dates_df, trips_per_day, on="date", how="left").fillna({"trip_count": 0})

# Export to CSV
export_path = r"P:\CarrerFoundy\Data Visualizations with Python\Achivement 2\trips_per_day_2022.csv"
trips_per_day_full.to_csv(export_path, index=False)

print("Trip counts for 2022 have been successfully exported.")

```

### 5.Created a line chart of bike trip counts and temperatures plotted on a dual axis. 

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Load the merged dataset
merged_path = r"P:\CarrerFoundy\Data Visualizations with Python\Achivement 2\merged_trips_weather_data_2022.csv"
merged_data = pd.read_csv(merged_path)

# Convert 'date' to datetime for plotting
merged_data["date"] = pd.to_datetime(merged_data["date"])

# Ensure numeric data types
merged_data["trip_count"] = pd.to_numeric(merged_data["trip_count"], errors="coerce")
merged_data["value"] = pd.to_numeric(merged_data["value"], errors="coerce")

# Normalize the temperature (Fixing the scaling issue)
merged_data["value"] = merged_data["value"] / 10  # Convert tenths of degrees Celsius to actual Celsius

# Create figure and axis
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot trip count on the primary y-axis (left)
ax1.plot(merged_data["date"], merged_data["trip_count"], color="blue", label="Bike Trips")
ax1.set_xlabel("Date")
ax1.set_ylabel("Trip Count", color="blue")
ax1.tick_params(axis="y", labelcolor="blue")

# Create a second y-axis (right) for temperature
ax2 = ax1.twinx()
ax2.plot(merged_data["date"], merged_data["value"], color="red", label="Temperature")
ax2.set_ylabel("Temperature (C)", color="red")
ax2.tick_params(axis="y", labelcolor="red")

# Format x-axis to show months
ax1.xaxis.set_major_locator(mdates.MonthLocator())  # Show each month
ax1.xaxis.set_major_formatter(mdates.DateFormatter('%b'))  # Format as 'Jan', 'Feb', etc.
plt.xticks(rotation=45)  # Rotate labels for readability

# Add grid lines for each month
ax1.grid(True, linestyle="--", alpha=0.6)

# Add title
plt.title("Bike Trip Counts and Temperature in 2022")

# Show the plot
plt.show()
```

### 6.  Briefly explained the code; used Matplotlib and uaed paradigm.

### Bike Trips vs. Temperature: Dual-Axis Line Chart  

This code visualizes daily bike trip counts and temperature for the year 2022 using Matplotlib.  
It uses the Object-Oriented (OO) paradigm of Matplotlib, which provides better control over the figure and axes.  

#### Key Steps in the Code:  
1. Load and Clean the data:  
   - Read `merged_trips_weather_data_2022.csv` and ensure the `date` column is in `datetime` format.  
   - Convert `trip_count` and `value` (temperature) to numeric types.  

2. Create a Dual-Axis Plot:  
   - Use `fig, ax1 = plt.subplots(figsize=(12, 6))` to initialize the figure and primary axis (`ax1`).  
   - Plot bike trip counts (blue) on `ax1`.  
   - Create a secondary y-axis (`ax2`) using `twinx()`, then plot temperature (red).  

3. Improve Readability:  
   - Use `mdates.MonthLocator()` and `mdates.DateFormatter('%b')` to display month names on the x-axis.  
   - Rotate month labels for clarity.  
   - Add grid lines to separate months.  

4. Display the Chart: 
   - Use `plt.show()` to render the visualization.  
   - The dual y-axis allows direct comparison of bike trips and temperature trends.  

#### Matplotlib Paradigm Used: 
This code follows the Object-Oriented (OO) approach, where:  
- The `figure` (`fig`) and `axes` (`ax1`, `ax2`) are explicitly created.  
- Each axis is independently controlled (e.g., setting labels, colors, and formatting).  
- The `twinx()` function is used to create a second y-axis, maintaining separate scales.  

This OO approach offers more flexibility than the state-based (pyplot) approach, making it ideal for complex visualizations like this one.
